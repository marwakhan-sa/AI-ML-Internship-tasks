# Task 1: Exploring and Visualizing a Simple Dataset (Iris Dataset)

Objective:
Learn how to load, inspect, and visualize a dataset to understand its trends, distributions, and relationships between features.

Dataset: 
Iris Dataset (loaded directly from seaborn library)

What will I do in this notebook:
- Load the Iris dataset using pandas and seaborn
- Check basic information: shape, column names, first few rows
- Get summary statistics using .info() and .describe()
- Create visualizations:
  - Scatter plot to see relationships between features
  - Histograms to show value distributions
  - Box plots to identify possible outliers

Tools used: 
- pandas  
- seaborn  
- matplotlib  

Goal:
Understand the structure of the data and discover patterns (e.g., how different flower species differ in measurements).

```python
import pandas as pd
import seaborn as sns

iris = sns.load_dataset('iris')

print("Shape of the dataset (rows, columns):", iris.shape)
print("\nColumn names:", iris.columns.tolist())
print("\nFirst 5 rows:")
print(iris.head())
```

```python
iris.info()

print("\nSummary statistics:")
print(iris.describe())
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.title('Sepal Length vs Sepal Width by Species')
plt.show()
```

```python
iris.hist(bins=20, figsize=(10, 8))
plt.suptitle('Distribution of Iris Features')
plt.show()
```

```python
sns.boxplot(data=iris)
plt.title('Box Plots of Iris Measurements')
plt.show()
```

 Results and Insights

- The Iris dataset contains 150 samples (rows) and 5 columns (4 numeric measurements + 1 species label).
- There are 3 species: setosa, versicolor, virginica.
- From the scatter plot:
  - Setosa is clearly separated from the other two species (especially in petal_length and petal_width).
  - Versicolor and virginica overlap a bit but can still be distinguished.
- Histograms show:
  - Sepal measurements are more normally distributed.
  - Petal_length and petal_width have more distinct groups (bimodal distribution).
- Box plots reveal:
  - A few potential outliers in sepal_width (mostly in setosa).
  - Petal measurements have tighter ranges for setosa compared to the others.

Key takeaway: Petal length and petal width are the most useful features for distinguishing between the three species. This dataset is clean, balanced, and a classic example for learning data exploration and visualization.
